```python
import pandas as pd 
import numpy as np
```

```python
wine=pd.read_csv('wine.data') #importing the dataset and assigning it to variable wine
```

```python
wine
```

```python
#Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
wine.drop(wine.columns[[0, 3, 6, 8,10,12,13]], axis='columns')
```

```python
#Assign the columns as below:
wine.columns=['Type','Alcohol','Malic acid','Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
```

```python
wine
```

```python
# Set the values of the first 3 rows from alcohol as NaN

wine['Alcohol'][:3]=np.nan
```

```python
wine
```

```python
#Set the value of the rows 3 and 4 of magnesium as NaN
wine['Magnesium'][2:4]=np.nan
```

```python
wine
```

```python
#Fill the value of NaN with the number 10 in alcohol and 100 in magnesium
wine['Alcohol'].replace(np.nan, 10,inplace=True)
wine['Magnesium'].replace(np.nan,100,inplace=True)
```

```python
wine
```

```python
#Count the number of missing values
wine.isnull().sum()
```

```python
#Create an array of 10 random numbers up until 10
random_array=np.random.randint(1,10,10)
random_array
```

```python
#Use random numbers you generated as an index and assign NaN value to each of cell.
wine.loc[random_array]=np.nan
```

```python
wine
```

```python
#How many missing values do we have?
wine.isnull().sum()
```

```python
#Delete the rows that contain missing values
wine.dropna(inplace = True)
```

```python
#Print only the non-null values in alcohol
print(wine.loc[wine['Alcohol'].notnull()])
```

```python
#Reset the index, so it starts with 0 again
wine.reset_index(drop = True, inplace = True)
wine
```
