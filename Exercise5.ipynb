```python
#importing req libraries
import pandas as pd
import numpy as np
```

```python
#importing the dataset and assigning to iris
iris=pd.read_csv('iris.data')
```

```python
#Create columns for the dataset
iris.columns=['sepal_length','sepal_width ','petal_length','petal_width','class']
```

```python
iris.head()
```

```python
#Is there any missing value in the dataframe
iris.isnull().sum()
```

```python
#set the values of the rows 10 to 29 of the column 'petal_length' to NaN
iris['petal_length'][9:29]=np.nan
```

```python
#substitute the NaN values to 1.0
iris['petal_length'].replace(np.nan,1.0,inplace=True)
```

```python
# delete the column class
del iris['class']
```

```python
#Set the first 3 rows as NaN
iris[:3]=np.nan
```
